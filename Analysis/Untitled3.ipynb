# Частотный анализ исторических шифров.

## Шифр Атбаш

Построение частотный гистограммы `шифра Атбаша`

```python
import matplotlib.pyplot as plt
import string
from collections import Counter
import itertools

RUSSIAN_ALP="абвгдежзийклмнопрстуфхцчшщъыьэюя"

def create_histogram(text: str, x: int, lab: str, alp=RUSSIAN_ALP, count_col=2) -> None:
    if ' ' in text:
        alphabet = list(alp.split())
        counter = Counter(text.lower().split())
    else:
        alphabet = list(alp)
        counter = Counter(text.lower())
    frequencies = [counter[letter] for letter in alphabet]
    plt.figure(figsize=(15,8))
    plt.subplot(count_col, 1, x)
    plt.bar(alphabet, frequencies)
    plt.xlabel('Буква')
    plt.ylabel('Частота')
    plt.title(lab)
    plt.show()

create_histogram("ТотЗПТктоложитсянадвастулаЗПТпадаетнаребраТЧК", 1, "Открытый текст")

create_histogram("нсншрнхнсфсщчноатяыэяонмфяшрнряыяънтяпъюпяних", 2, "Шифр Атбаш")

```

Размытия статистических характеристик в данном шифре не происходит, т.к. буквы алфавита только лишь меняются ролями. В итоге использования данного шифра получаем зеркальное отражение статистических характеристик в виду особенности алгоритма замены.

## Шифр Плейфера

```python
create_histogram("ТотЗПТктоложитсянадвастулаЗПТпадаетнаребраТЧК", 1, "Открытый текст")

alp = itertools.product(RUSSIAN_ALP, repeat=2)
tmp = []
for i in alp:
    tmp.append(''.join(i))
en_message = "то тз пт кт ол ож ит ся на дв ас ту ла зп тп ад ае тн ар еб ра тч кф"

alp = list(filter(lambda x: x in en_message, tmp))
alp = ' '.join(alp)

create_histogram("то тз пт кт ол ож ит ся на дв ас ту ла зп тп ад ае тн ар еб ра тч кф", 2, "Шифр Плейфера", alp)
create_histogram("пбсисклкапжптштюарнздлкшфгжсксрбргкброирорсшлу", 2, "Шифр Плейфера")
```

Шифр Плейфера является шифром биграммной замены, поэтому если анализировать его биграммы на данном коротком открытом тексте, гистограмма асболютно ровная
Если анализовать его по одному символу, то шифр размывает статистические характеристики, но присутствуют "пики"

## Одноразовый блокнот Шеннона

ГЕНЕРАТОР ГАММЫ


```python
create_histogram("ТотЗПТктоложитсянадвастулаЗПТпадаетнаребраТЧК", 1, "Открытый текст")

alp = list(map(str, list(range(1, 32))))
for i in range(len(alp)):
    while len(alp[i]) != 2:
        alp[i] = '0' + alp[i]
alp =  ' '.join(alp)
create_histogram("24 01 26 20 23 25 14 11 12 10 14 19 22 17 15 48 21 31 28 15 22 02 06 28 31 23 25 20 28 02 12 05 10 08 26 18 06 27 03 26 18 07 1", 2, "одноразовый блокнот шеннона", alp)
```

## Шифр ElGamal

```python
create_histogram("ТотЗПТктоложитсянадвастулаЗПТпадаетнаребраТЧК", 1, "Открытый текст")
en_message = "4103 1114 0405 1139 1135 0405 1116 1112 1114 0419 4125 4126 0425 1112 1134 1127 0415 1121 0433 0437 1121 4130 0405 0403 1137 1121 1139 1135 4103 0411 0441 1119 0441 1140 0111 2113 6044 1111 3114 0114 2411 4112 1040 5414 0042 1"
alp = list(map(str, list(range(1, 10000))))
for i in range(len(alp)):
    while len(alp[i]) != 4:
        alp[i] = '0' + alp[i]
alp = ' '.join(list(filter(lambda x: x in en_message, alp)))

create_histogram(en_message, 2, "ELGAMAL", alp)
```

Шифр ElGamal показывает наилучшее размытие статистических характеристик. График практически ровный.
